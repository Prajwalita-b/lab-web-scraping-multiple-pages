```python
df = pd.DataFrame()

for start in range(1,5):
    rs = requests.get(f'https://www.last.fm/tag/rolling+stone+500+greatest+songs+of+all+time/tracks?page={start}')
    soup = BeautifulSoup(rs.content, 'html.parser')
    data_s = []
    containers = soup.find_all('td', attrs = {'class': 'chartlist-name' })
    for container in containers:
        song = container.find('a').get_text(strip = True)
        data_s.append((song))
    pd.DataFrame(data_s, columns = ['song'])
    data_a = []
    container_a = soup.find_all('td', attrs = {'class': 'chartlist-artist' })
    for containera in container_a:
        artist = containera.find('a').get_text(strip = True)
        data_a.append((artist))
    pd.DataFrame(data_a, columns = ['artist'])
    
    new_df = pd.DataFrame(zip(data_a, data_s), columns=["artist name", "song name"])   
    df = pd.concat([df, new_df])

df = df.reset_index()
df.index = df.index + 1

df
```
